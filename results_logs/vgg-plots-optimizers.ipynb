```python
import sys
import  pandas as pd
import matplotlib.pyplot as plt

sys.path.insert(0,'/content/drive/MyDrive/ColabNotebooks/')
```

```python
SGDv3= pd.read_csv("/content/drive/MyDrive/ColabNotebooks/results_logs/learning-history/75/vgg16/vgg-SGD.txt", sep =",")
```

```python
SWAv3 = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/results_logs/learning-history/75/vgg16/vgg-SWA.txt", sep =",")
```

```python
SAMv3 = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/results_logs/learning-history/75/vgg16/vgg-SAM.txt", sep =",")
```

```python
GSAMv3= pd.read_csv("/content/drive/MyDrive/ColabNotebooks/results_logs/learning-history/75/vgg16/vgg-GSAM.txt", sep =",")
```

```python
LIONv3= pd.read_csv("/content/drive/MyDrive/ColabNotebooks/results_logs/learning-history/75/vgg16/vgg-Lion.txt", sep =",")
```

```python
LIONv3['accuracy_train'] = LIONv3['accuracy_train'] * 100
SAMv3['accuracy_train'] = SAMv3['accuracy_train'] * 100
SWAv3['accuracy_train'] = SWAv3['accuracy_train'] * 100
SGDv3['accuracy_train'] = SGDv3['accuracy_train'] * 100
```

```python
plt.figure(figsize=(10, 6))

plt.plot(SWAv3['epoch'], SWAv3['accuracy_train'], label='SWA')
plt.plot(SGDv3['epoch'], SGDv3['accuracy_train'], label='SGD')
plt.plot(SAMv3['epoch'], SAMv3['accuracy_train'], label='SAM')
plt.plot(GSAMv3['epoch'], GSAMv3['accuracy_train'], label='GSAM')
plt.plot(LIONv3['epoch'], LIONv3['accuracy_train'], label='LION')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# plt.ylim(0, 100)

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(SWAv3['epoch'], SWAv3['accuracy_valid'], label='SWA')
plt.plot(SAMv3['epoch'], SAMv3['accuracy_valid'], label='SAM')
plt.plot(SGDv3['epoch'], SGDv3['accuracy_valid'], label='SGD')
plt.plot(GSAMv3['epoch'], GSAMv3['accuracy_valid'], label='GSAM')
plt.plot(LIONv3['epoch'], LIONv3['accuracy_valid'], label='LION')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# plt.ylim(0, 100)

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(SWAv3['epoch'], SWAv3['loss_train'], label='SWA')
plt.plot(SAMv3['epoch'], SAMv3['loss_train'], label='SAM')
plt.plot(SGDv3['epoch'], SGDv3['loss_train'], label='SGD')
plt.plot(GSAMv3['epoch'], GSAMv3['loss_train'], label='GSAM')
plt.plot(LIONv3['epoch'], LIONv3['loss_train'], label='LION')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# plt.ylim(0, 100)

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(SWAv3['epoch'], SWAv3['loss_valid'], label='SWA')
plt.plot(SAMv3['epoch'], SAMv3['loss_valid'], label='SAM')
plt.plot(SGDv3['epoch'], SGDv3['loss_valid'], label='SGD')
plt.plot(GSAMv3['epoch'], GSAMv3['loss_valid'], label='GSAM')
plt.plot(LIONv3['epoch'], LIONv3['loss_valid'], label='LION')


plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# plt.ylim(0, 100)

plt.grid(True)
plt.show()
```
