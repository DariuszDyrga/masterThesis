```python
import sys
import  pandas as pd
import matplotlib.pyplot as plt

sys.path.insert(0,'/content/drive/MyDrive/ColabNotebooks/')
```

```python
vgg= pd.read_csv("/content/drive/MyDrive/ColabNotebooks/results_logs/learning-history/75/vgg16/vgg-GSAM.txt", sep =",")
```

```python
mobilenet= pd.read_csv("/content/drive/MyDrive/ColabNotebooks/results_logs/learning-history/75/mobilenet/mobilenet-GSAM.txt", sep =",")
```

```python
inceptionv3= pd.read_csv("/content/drive/MyDrive/ColabNotebooks/results_logs/learning-history/75/inceptionv3/inception-GSAM.txt", sep =",")
```

```python
widerestnet =  pd.read_csv("/content/drive/MyDrive/ColabNotebooks/results_logs/learning-history/75/widerestnet/widerestnet-GSAM.txt", sep =",")
```

```python
vit = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/results_logs/learning-history/75/vit224/vit-gsam-224.txt")
```

```python
plt.figure(figsize=(10, 6))

plt.plot(vit['epoch'], vit['accuracy_train'], label='ViT')
plt.plot(vgg['epoch'], vgg['accuracy_train'], label='VGG')
plt.plot(mobilenet['epoch'], mobilenet['accuracy_train'], label='MobileNet')
plt.plot(inceptionv3['epoch'], inceptionv3['accuracy_train'], label='InceptionV3')
plt.plot(widerestnet['epoch'], widerestnet['accuracy_train'], label='WideRestNet')

plt.xlabel('Epoch')
plt.ylabel('Accuracy train')
plt.legend()

# plt.ylim(0, 100)

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(vit['epoch'], vit['accuracy_valid'], label='ViT')
plt.plot(vgg['epoch'], vgg['accuracy_valid'], label='VGG')
plt.plot(mobilenet['epoch'], mobilenet['accuracy_valid'], label='MobileNet')
plt.plot(inceptionv3['epoch'], inceptionv3['accuracy_valid'], label='InceptionV3')
plt.plot(widerestnet['epoch'], widerestnet['accuracy_valid'], label='WideRestNet')

plt.xlabel('Epoch')
plt.ylabel('Accuracy valid')
plt.legend()

# plt.ylim(0, 100)

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(vit['epoch'], vit['loss_train'], label='ViT')
plt.plot(vgg['epoch'], vgg['loss_train'], label='VGG')
plt.plot(mobilenet['epoch'], mobilenet['loss_train'], label='MobileNet')
plt.plot(inceptionv3['epoch'], inceptionv3['loss_train'], label='InceptionV3')
plt.plot(widerestnet['epoch'], widerestnet['loss_train'], label='WideRestNet')

plt.xlabel('Epoch')
plt.ylabel('Loss train')
plt.legend()

# plt.ylim(0, 100)

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(vit['epoch'], vit['loss_valid'], label='ViT')
plt.plot(vgg['epoch'], vgg['loss_valid'], label='VGG')
plt.plot(mobilenet['epoch'], mobilenet['loss_valid'], label='MobileNet')
plt.plot(inceptionv3['epoch'], inceptionv3['loss_valid'], label='InceptionV3')
plt.plot(widerestnet['epoch'], widerestnet['loss_valid'], label='WideRestNet')

plt.xlabel('Epoch')
plt.ylabel('Loss valid')
plt.legend()

# plt.ylim(0, 100)

plt.grid(True)
plt.show()
```
